## Análisis de la Inversión Bruta Fija Privada y Pública

```python
df1 = pd.read_excel('Tarea3_DE.xlsx', sheet_name='Hoja2')

display(df1.head())
```

```python
def convert_quarterly_date(date_str):
    """Converts a string in 'Tqyy' format to a datetime object."""
    quarter = int(date_str[1])
    year = 2000 + int(date_str[2:])
    month = (quarter - 1) * 3 + 1
    return pd.to_datetime(f'{year}-{month:02d}-01')

df1['Fecha'] = df1['Fecha'].apply(convert_quarterly_date)
df1.set_index('Fecha', inplace=True)

display(df1.head())
```

## Análisis de Regresión Lineal Simple

```python
import statsmodels.api as sm
```

```python
# Define dependent variable
y = df['GC']

# Define independent variables
X = df[['GC - Remuneraciones', 'IBF - Publica']]

# Add a constant to the independent variables
X = sm.add_constant(X)

# Fit the linear regression model
model = sm.OLS(y, X)
results = model.fit()

# Print the summary of the regression results
print(results.summary())
```

1. Información General del Modelo:

- Dep. Variable: GC, La variable dependiente es el "Gasto Corriente" (GC), lo que significa que el modelo intenta explicar y predecir el Gasto Corriente en función de las variables independientes.

- Model: OLS. Se utilizó el método de Mínimos Cuadrados Ordinarios, que busca minimizar la suma de los cuadrados de los errores entre los valores observados y los predichos.

2. Bondad de Ajuste del Modelo:

- R-squared: 0.941. Este es un valor muy alto. Significa que el 94.1% de la variabilidad en el Gasto Corriente (GC) es explicada por las variables independientes (Gasto Corriente en Remuneraciones e Inversión Bruta Fija Pública) en el modelo. Esto indica un excelente ajuste del modelo a los datos.

- Adj. R-squared: 0.940. El R-cuadrado ajustado es ligeramente inferior al R-cuadrado, lo cual es normal. Confirma el alto poder explicativo del modelo, ajustado por el número de predictores.

- F-statistic: 1828. Este es el valor del estadístico F para la prueba de significancia global del modelo.

- Prob (F-statistic): 6.49e-142. Este valor (esencialmente 0) es extremadamente pequeño. Dado que es mucho menor que cualquier nivel de significancia común (e.g., 0.05 o 0.01), podemos rechazar la hipótesis nula de que todos los coeficientes de las variables independientes son cero. En otras palabras, el modelo en su conjunto es estadísticamente significativo y tiene poder predictivo.

3. Coeficientes de Regresión e Interpretación:

- const (Intercepción): 467.8997. Interpretación: Cuando "GC - Remuneraciones" y "IBF - Pública" son cero, el Gasto Corriente (GC) estimado es de aproximadamente 467.90 millones de S/.

- P>|t|: 0.002. Dado que el p-valor (0.002) es menor que 0.05 (e incluso 0.01), la constante es estadísticamente significativa.

- GC - Remuneraciones: 1.9993. Interpretación: Por cada aumento de 1 millón de S/. en el "Gasto Corriente en Remuneraciones", el "Gasto Corriente" (GC) aumenta en aproximadamente 1.9993 millones de S/., manteniendo constante la "IBF - Pública".

```python
from statsmodels.tsa.stattools import adfuller

def run_adf_test(series, name):
    """Performs and prints the Augmented Dickey-Fuller test results."""
    print(f'Results for Augmented Dickey-Fuller Test on "{name}":')
    result = adfuller(series.dropna())
    print(f'ADF Statistic: {result[0]}')
    print(f'p-value: {result[1]}')
    print('Critical Values:')
    for key, value in result[4].items():
        print(f'\t{key}: {value}')

    if result[1] <= 0.05:
        print(f'The series "{name}" is likely stationary (p-value <= 0.05).\n')
    else:
        print(f'The series "{name}" is likely non-stationary (p-value > 0.05).\n')

run_adf_test(df['GC'], 'GC')
run_adf_test(df['IBF - Publica'], 'IBF - Publica')
```

**Análisis**:

La prueba ADF es un test estadístico que se utiliza para determinar si una serie de tiempo tiene una raíz unitaria, lo que esencialmente nos dice si la serie es estacionaria o no estacionaria. Una serie de tiempo estacionaria tiene propiedades estadísticas constantes a lo largo del tiempo (media, varianza y autocorrelación), mientras que una no estacionaria no las tiene (a menudo presenta una tendencia).

Análisis de la Prueba ADF para "GC" (Gasto Corriente):

Estadístico ADF: 0.26458

Valor p: 0.97566

Valores Críticos:

1%: -3.46057

10%: -2.57385

Interpretación para "GC":

El valor p (0.97566) es significativamente mayor que los niveles de significancia comunes (por ejemplo, 0.01, 0.05, 0.10). Además, el estadístico ADF (0.26458) es mucho mayor (menos negativo) que todos los valores críticos.

Conclusión para "GC": Basándonos en estos resultados, no podemos rechazar la hipótesis nula de la prueba ADF. La hipótesis nula establece que existe una raíz unitaria, lo que implica que la serie es no estacionaria. Por lo tanto, la serie "GC" es muy probablemente no estacionaria. Esto significa que presenta una estructura dependiente del tiempo, como una tendencia, lo que puede conducir a regresiones espurias si no se maneja adecuadamente.

## Evaluación del Performance del Modelo



```python
p_values = range(0, 3)
d_values = range(0, 3)
q_values = range(0, 3)

print(f"p_values: {list(p_values)}")
print(f"d_values: {list(d_values)}")
print(f"q_values: {list(q_values)}")
```

```python
import warnings
warnings.filterwarnings('ignore')

best_aic_gc = []
best_aic_ibf = []

for p in p_values:
    for d in d_values:
        for q in q_values:
            order = (p, d, q)
            try:
                model_gc = ARIMA(df['GC'], order=order)
                model_fit_gc = model_gc.fit()
                best_aic_gc.append({'order': order, 'aic': model_fit_gc.aic})

                model_ibf = ARIMA(df['IBF - Publica'], order=order)
                model_fit_ibf = model_ibf.fit()
                best_aic_ibf.append({'order': order, 'aic': model_fit_ibf.aic})

            except ValueError:
                print(f'ValueError for order: {order}')

print("AIC values for GC:")
print(best_aic_gc)

print("\nAIC values for IBF - Publica:")
print(best_aic_ibf)
```

## Identificando el Mejor Modelo


```python
best_gc_order = min(best_aic_gc, key=lambda x: x['aic'])
best_ibf_order = min(best_aic_ibf, key=lambda x: x['aic'])

print(f"Best ARIMA order for GC: {best_gc_order['order']} with AIC: {best_gc_order['aic']}")
print(f"Best ARIMA order for IBF - Publica: {best_ibf_order['order']} with AIC: {best_ibf_order['aic']}")
```

```python
model_gc_best = ARIMA(df['GC'], order=best_gc_order['order'])
results_gc_best = model_gc_best.fit()

model_ibf_best = ARIMA(df['IBF - Publica'], order=best_ibf_order['order'])
results_ibf_best = model_ibf_best.fit()

print("Best ARIMA model for GC summary:")
print(results_gc_best.summary())

print("\nBest ARIMA model for IBF - Publica summary:")
print(results_ibf_best.summary())
```
